```python
from PIL import Image, ImageDraw
from IPython.display import display
import numpy as np


# Create a white canvas
width, height = 400, 400
img = Image.new('RGB', (width, height), 'white')
draw = ImageDraw.Draw(img)

# Draw a rectangle (black)
draw.rectangle([50, 50, 150, 150], fill='black')

# Draw a circle (black)
draw.ellipse([200, 200, 300, 300], fill='black')
```

```python
# Create mask for white pixels (background)
# Convert to numpy array for masking
img_array = np.array(img)

# Create mask for white pixels (background)
mask = (img_array == [255, 255, 255]).all(axis=2)
display(img)
```

```python

# Change white background to red, leaving shapes unchanged
img_array[mask] = [255, 0, 0]  # Red color

# Convert back to image
img = Image.fromarray(img_array)

# Display in Jupyter
display(img)
```
