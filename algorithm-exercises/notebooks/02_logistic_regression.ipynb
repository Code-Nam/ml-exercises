```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Score d'examen -> admission (0 = non, 1 = oui)
X = np.array([[40], [50], [60], [65], [70], [75], [80], [85], [90]])
Y = np.array([0, 0, 0, 0, 1, 1, 1, 1, 1])
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
# Plot the data points
plt.scatter(X, Y, color='blue', label='Data points')

# Predict probabilities using the trained model
# Generate a range of exam scores for smooth curve
X_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)

# Predict probabilities using the trained model
Y_pred = model.predict_proba(X_range)[:, 1]

# Plot the logistic curve
plt.plot(X_range, Y_pred, color='red', label='Logistic Regression Curve')

# Plot the predicted point for score 70
X_single = np.array([[70]])
Y_single_pred = model.predict_proba(X_single)[:, 1]
plt.scatter(X_single, Y_single_pred, color='green', label='Predicted Probability for Score 70')

# Add labels and legend
plt.xlabel('Exam Score')
plt.ylabel('Probability of Admission')
plt.title('Logistic Regression: Exam Score vs Admission')
plt.legend()
plt.grid(True)
plt.show()
```
